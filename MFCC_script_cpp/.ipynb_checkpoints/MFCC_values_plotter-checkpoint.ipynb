```python
import pandas as pd
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a heatmap
# Read CSV file
df = pd.read_csv('MFCC_quantized_values.csv', header=None)

# Extract MFCC values as a NumPy array
mfcc_values = df.values

plt.figure(figsize=(10, 6))
plt.imshow(mfcc_values, cmap='viridis', aspect='auto', interpolation = "none")

# Add labels and title
plt.xlabel('MFCC Coeffs')
plt.ylabel('Frames')
plt.title('MFCC Heatmap')

# Show colorbar
plt.colorbar(label='MFCC Values')

# Show the plot
plt.show()
```
